# Polynomial Rregression

# Importing the libraraies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

# Importing the dataset

```python
file_path = 'Pract.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
X = data[['open', 'high', 'low', 'volume']]  # Independent variables
y = data['close']  # Dependent variable
```

```python
# Create polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

# Splitting the dataset into Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Training the model on polynomial regression

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

# Predicting the values

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

# Evaluating the model

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

# Visualizing the model

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot Actual vs. Predicted

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2)
plt.title('Actual vs Predicted Close Prices')
plt.xlabel('Actual Close Prices')
plt.ylabel('Predicted Close Prices')
plt.grid(True)
plt.show()
```

# Plot Residuals

```python
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.title('Residual Plot')
plt.xlabel('Predicted Close Prices')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
